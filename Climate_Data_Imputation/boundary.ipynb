```python
import geopandas as gpd
import pandas as pd
```

### we can download geographic boundary data from census.gov
For example, county boundary data can be found : https://www.census.gov/geographies/mapping-files/time-series/geo/carto-boundary-file.html

```python
# Load the Shapefile
gdf = gpd.read_file('CountyBoundary/cb_2018_us_county_500k.shp')

# Convert the GeoDataFrame to a DataFrame
county = pd.DataFrame(gdf)
```

```python
county
```

```python
county.to_csv('county_boundary.csv')
```
